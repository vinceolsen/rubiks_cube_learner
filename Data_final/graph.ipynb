```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
etas = [i / 10 for i in range(10)]
files = [np.loadtxt(f'eta_{eta}_test.txt', dtype=int) for eta in etas]
scores = [np.count_nonzero(files[i] < 20) for i in range(10)]

print(scores)
```

```python
gammas = [i / 10 for i in range(10)]
gammafiles = [np.loadtxt(f'eta_0.3_gamma_{gamma}_test.txt', dtype=int) for gamma in gammas]
gammaScores = [np.count_nonzero(gammafiles[i] < 20) for i in range(10)]

print(gammaScores)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.set_size_inches(10, 5)
# ax = fig.add_axes([0,0,1,1])
etaString = [str(eta) for eta in etas]
percentScore = [i / 10 for i in scores]
ax1.bar(etaString, percentScore)
ax1.set_xlabel('Eta')
ax1.set_ylabel('Successful Tests (%)')
ax1.set_title('Percent of Successful Tests for Eta Values')

gammaString = [str(gamma) for gamma in gammas]
percentGammaScore = [i / 10 for i in gammaScores]
ax2.bar(gammaString, percentGammaScore)
ax2.set_xlabel('Gamma')
ax2.set_ylabel('Successful Tests (%)')
ax2.set_title('Percent of Successful Tests for Gamma Values')

plt.show()
```

```python
depths = [i for i in range(5)]
depthFiles = [np.loadtxt(f'eta_0.3_gamma_0.9_testdepth_{depth}_test.txt', dtype=int) for depth in depths]
depthScores = [np.count_nonzero(depthFiles[i] < 20) for i in range(5)]
```

```python
fig, ax1 = plt.subplots()
depthString = [str(depth) for depth in depths]
percentDepthScore = [i / 10 for i in depthScores]
ax1.bar(depthString, percentDepthScore)
ax1.set_title('Percent of Successful Tests for Depth from ADI Training')
plt.xlabel('Scramble Length Past ADI')
plt.ylabel('Successful Tests (%)')
plt.show()
```

```python
depths = [i for i in range(10)]
depthFiles = [np.loadtxt(f'eta_0.3_gamma_0.9_testdepth_{depth}_trainedToo_test.txt', dtype=int) for depth in depths]
depthScores = [np.count_nonzero(depthFiles[i] < 20) for i in range(10)]
```

```python
# fig, ax1 = plt.subplots()
depthString = [str(depth) for depth in depths]
depth = [str(i) for i in range(4, 14)]
percentDepthScore = [i / 10 for i in depthScores]
plt.barh(depth, percentDepthScore)

for index, value in enumerate(percentDepthScore):
    plt.text(value, index, str(value))

plt.title('Percent of Successful Tests for ADI Training Depth = 4')
plt.ylabel('Depth from Goal State')
plt.xlabel('Successful Tests (%)')
plt.show()
```
